```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
%matplotlib inline


data = load_boston()
X = data.data
y = data.target
df = pd.DataFrame(X, columns = data.feature_names)
df["MEDV"] = y
n, m = df.shape # n: number of rows, m: number of columns
```
