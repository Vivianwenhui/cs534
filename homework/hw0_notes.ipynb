```python
import numpy as np
import pandas as pd
import seaborn as sns
df = pd.read_csv("hw0_p1.csv")
def my_guess(x1, x2):
    return x2 * 1.5 + x1 * x2 + np.abs(x2)* 0.5

df["error"] = df["y"] - my_guess(df["x1"], df["x2"])
sns.pairplot(df)
```

# P2

```python
import numpy as np
import pandas as pd
df = pd.read_csv("AAPL.csv")
df["Close_yesterday"] = df["Close"].shift(1)
df["Year"] = pd.DatetimeIndex(df["Date"]).year
df_2014 = df.loc[df["Year"]==2014,:]
daily_ret = np.mean(df["Close"]/df["Close_yesterday"])
daily_ret_2014 = np.mean(df_2014["Close"]/df_2014["Close_yesterday"])
print(f"Daily Return: {daily_ret}\nDaily Return (2014): {daily_ret_2014}")
```

# P3

```python
import numpy as np

choco_to_orange = np.array([[0.7, 0], [0.3, 1]])
orange_to_choco = np.array([[1, 0.2], [0, 0.8]])
one_cycle = np.dot(orange_to_choco, choco_to_orange)
state = np.array([1, 1])
data = [[0] + state.tolist()]
for i in range(100):
    state = np.dot(one_cycle, state)
    data.append([i+1] + state.tolist())
df = pd.DataFrame(data, columns=['cycle', 'choco', 'orange'])
df.set_index("cycle").plot.line()
```
